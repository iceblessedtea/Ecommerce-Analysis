# Langkah 5: Menentukan Objek Data

Dataset e-commerce ini terdiri dari empat entitas utama yaitu Customer, Product, Transaction, dan Review.
Entitas-entitas tersebut saling terhubung untuk membentuk proses bisnis yang utuh, dimulai dari pendaftaran pelanggan, transaksi pembelian produk, hingga pemberian ulasan dan rating.

Hubungan antar-entitas bersifat one-to-many, seperti satu pelanggan dapat melakukan banyak transaksi, dan satu produk dapat menerima banyak ulasan.
Struktur ini memungkinkan analisis perilaku pelanggan, performa produk, serta evaluasi kepuasan pelanggan melalui analisis review dan rating.

Berdasarkan analisis singkat mengenai dataset, maka ditemukan bahwa :

###  Tujuan
Menentukan entitas utama dalam dataset e-commerce, atribut kunci dari setiap entitas, serta hubungan antar entitas untuk mendukung analisis data dan pembuatan dashboard Business Intelligence.

---

###  Identifikasi Entitas Data
Berdasarkan struktur dataset, terdapat empat entitas utama yang mewakili proses bisnis e-commerce:

| Entitas | Deskripsi | Contoh Kolom |
|----------|------------|---------------|
| **Customer** | Menyimpan data pelanggan yang melakukan pembelian. | `customer_id`, `first_name`, `last_name`, `gender`, `age_group`, `country`, `signup_date` |
| **Product** | Menyimpan informasi barang yang dijual. | `product_id`, `product_name`, `category`, `unit_price` |
| **Transaction (Order)** | Mencatat setiap transaksi pembelian pelanggan. | `order_id`, `customer_id`, `product_id`, `order_date`, `quantity`, `payment_method`, `order_status` |
| **Review** | Menyimpan umpan balik pelanggan setelah pembelian. | `review_id`, `order_id`, `product_id`, `review_text`, `rating`, `review_date` |

---

###  Atribut Kunci (Key Attributes)

| Entitas | Primary Key | Foreign Key (Relasi ke) |
|----------|--------------|--------------------------|
| Customer | `customer_id` | Digunakan di `Transaction` |
| Product | `product_id` | Digunakan di `Transaction` dan `Review` |
| Transaction | `order_id` | Dihubungkan ke `Customer` dan `Product` |
| Review | `review_id` | Dihubungkan ke `order_id` (Transaction) |

---

### Hubungan Antar Entitas

| Hubungan | Jenis Relasi | Deskripsi |
|-----------|---------------|------------|
| **Customer → Transaction** | 1 : n | Satu pelanggan dapat melakukan banyak transaksi. |
| **Product → Transaction** | 1 : n | Satu produk dapat dibeli oleh banyak pelanggan. |
| **Transaction → Review** | 1 : 1 atau 1 : n | Satu transaksi dapat memiliki satu atau beberapa review. |
| **Product → Review** | 1 : n | Satu produk dapat menerima banyak ulasan pelanggan. |

---

### Alur Bisnis Data

1. **Customer** melakukan registrasi (`signup_date`).  
2. **Customer** melakukan **Transaction** terhadap satu atau beberapa **Product**.  
3. Setiap transaksi memiliki informasi seperti `order_date`, `quantity`, `payment_method`, dan `order_status`.  
4. Setelah transaksi selesai, pelanggan memberikan **Review** (`rating`, `review_text`, `review_date`).  
5. Data review ini kemudian digunakan untuk menilai **kualitas produk** dan **kepuasan pelanggan**.

```python
import pandas as pd

# --- 1. MEMUAT DATA ---
try:
    df = pd.read_csv('cleandata.csv')
except FileNotFoundError:
    print("Error: File 'cleandata.csv' tidak ditemukan. Pastikan file berada di direktori yang sama.")
    # Hentikan eksekusi jika file tidak ditemukan
    exit()

# Persiapan data: Fokus pada rating dan review_text, ubah teks ke huruf kecil
df_inconsistent = df[['rating', 'review_text']].copy()
df_inconsistent['review_text'] = df_inconsistent['review_text'].astype(str).str.lower()

# --- 2. DEFINISI KATA KUNCI SENTIMEN ---
# Kata kunci untuk mendeteksi sentimen yang bertentangan
# Contoh: 'good' atau 'very good'
positive_keywords = r'good|very good|excellent|great'
# Contoh: 'bad', 'very bad', atau 'average'
negative_keywords = r'bad|very bad|poor|terrible|average'

# --- 3. FILTER INKONSISTENSI UTAMA ---

## Filter A: Rating Rendah (1 atau 2) TAPI Teks Positif
# Kasus: Rating 1/2, Teks "very good"
filter_A = (df_inconsistent['rating'].isin([1, 2])) & \
           (df_inconsistent['review_text'].str.contains(positive_keywords, regex=True))

## Filter B: Rating Tinggi (4 atau 5) TAPI Teks Negatif/Netral
# Kasus: Rating 4/5, Teks "bad"
filter_B = (df_inconsistent['rating'].isin([4, 5])) & \
           (df_inconsistent['review_text'].str.contains(negative_keywords, regex=True))

# Gabungkan hasil kedua filter dan hapus duplikat
inconsistent_rows = df_inconsistent[filter_A | filter_B].drop_duplicates()

# --- 4. TAMPILKAN HASIL ---
print("## Pengecekan Rating dan Review Text yang Tidak Sesuai")
print("----------------------------------------------------------")
print(f"Total baris inkonsisten yang ditemukan: {len(inconsistent_rows)} baris.")

if not inconsistent_rows.empty:
    print("\nContoh 20 Baris Inkonsisten Pertama:")
    # Menggunakan to_string() untuk format teks umum
    print(inconsistent_rows.head(20).to_string(index=True))
else:
    print("\nTidak ada inkonsistensi yang ditemukan berdasarkan kriteria kata kunci yang ketat.")
```

# Klasifikasi payment method 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
target_kolom = 'payment_method'
fitur_numerik = ['quantity', 'unit_price', 'total_sales', 'rating']
fitur_kategori = ['gender', 'age_group', 'country', 'category', 'Order_Year']
```

```python
kolom_dipilih = [target_kolom] + fitur_numerik + fitur_kategori
```

```python
df_klasifikasi = df[kolom_dipilih].copy()
df_klasifikasi.head()
```

```python
for col in fitur_numerik:
    df_klasifikasi[col].fillna(df_klasifikasi[col].median(), inplace=True)
for col in fitur_kategori:
    df_klasifikasi[col].fillna(df_klasifikasi[col].mode()[0], inplace=True)
```

```python
le = LabelEncoder()
df_klasifikasi['payment_method_encoded'] = le.fit_transform(df_klasifikasi[target_kolom])
print(f"Kelas Payment Method yang terdeteksi: {le.classes_}")
```

```python
# Pisahkan Fitur (X) dan Target (Y)
X = df_klasifikasi[fitur_numerik + fitur_kategori]
Y = df_klasifikasi['payment_method_encoded']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print(f"Data Training: {len(X_train)} baris")
print(f"Data Testing: {len(X_test)} baris")
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        # Transformer 1: Scaling untuk kolom numerik
        ('num', StandardScaler(), fitur_numerik),
        # Transformer 2: One-Hot Encoding untuk kolom kategori
        ('cat', OneHotEncoder(handle_unknown='ignore'), fitur_kategori)
    ])
```

```python
models = {
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1),
    'LogisticRegression': LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000, random_state=42),
    'SupportVectorMachine': SVC(random_state=42),
    'KNearestNeighbors': KNeighborsClassifier(n_neighbors=5)
}
```

```python
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1))
])
```

```python
model_pipeline.fit(X_train, Y_train)
```

```python
Y_pred = model_pipeline.predict(X_test)
```

```python
print("## Hasil Evaluasi Model Random Forest")
print(classification_report(Y_test, Y_pred, target_names=le.classes_))
```

```python
cm = confusion_matrix(Y_test, Y_pred)
print("## Confusion Matrix")
print(cm)
```

```python
# Import library model tambahan
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

# --- 1. DAFTAR MODEL YANG AKAN DICOBA ---
models = {
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1),
    'LogisticRegression': LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000, random_state=42),
    'SupportVectorMachine': SVC(random_state=42),
    'KNearestNeighbors': KNeighborsClassifier(n_neighbors=5)
}

results = {}

# --- 2. PELATIHAN DAN EVALUASI SETIAP MODEL ---
print("--- Memulai Pelatihan dan Evaluasi Multimodel ---")

for name, classifier in models.items():
    print(f"\nMelatih Model: {name}...")
    
    # Membuat pipeline untuk setiap model
    model_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', classifier)
    ])
    
    # Melatih model
    model_pipeline.fit(X_train, Y_train)
    
    # Memprediksi dan Evaluasi
    Y_pred = model_pipeline.predict(X_test)
    
    # Menyimpan hasil evaluasi (Classification Report)
    report = classification_report(Y_test, Y_pred, target_names=le.classes_, output_dict=True)
    
    # Ambil Akurasi keseluruhan (macro avg F1-score atau accuracy)
    accuracy = report['accuracy'] 
    
    results[name] = {
        'accuracy': accuracy,
        'report': report
    }
    
    print(f"Model {name} selesai. Akurasi: {accuracy:.4f}")

print("-------------------------------------------------")

# --- 3. MENAMPILKAN HASIL AKHIR DAN PERBANDINGAN ---

print("## Rekapitulasi Akurasi Model Klasifikasi")
recap_df = pd.DataFrame({
    'Model': list(results.keys()),
    'Akurasi': [results[name]['accuracy'] for name in results.keys()]
}).sort_values(by='Akurasi', ascending=False).reset_index(drop=True)



print("\n## Hasil Detail (Classification Report) dari Model Terbaik")
best_model_name = recap_df.iloc[0]['Model']
best_report = results[best_model_name]['report']

print(f"\nModel Terbaik: **{best_model_name}** (Akurasi: {best_report['accuracy']:.4f})")
print(classification_report(Y_test, model_pipeline.predict(X_test), target_names=le.classes_))
```
